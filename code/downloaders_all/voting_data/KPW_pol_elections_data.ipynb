```python
import pandas as pd
from pathlib import Path

raw_path = Path(r"C:\Users\jarem\OneDrive - London School of Economics\YEAR 2\1. Policy paper\policy-paper-repo\data\clean\outcome\Elections\raw")

files_to_check = {
    2001: "sejm2001-lis-gm.xls",
    2007: "sejm2007-gm-listy.xls", 
    2011: "2011-gl-lis-gm.xls",
    "2005a": "1456225675_36795.xls",
    "2005b": "1456225675_36796.xls",
}

for year, filename in files_to_check.items():
    folder = "2005" if "2005" in str(year) else str(year)
    path = raw_path / folder / filename
    print(f"\n{'='*60}")
    print(f"  {year}: {filename}")
    print(f"{'='*60}")
    try:
        df = pd.read_excel(path, nrows=3)
        print(f"  Shape (full): checking cols only")
        print(f"  Columns ({len(df.columns)}):")
        for c in df.columns:
            print(f"    {repr(c)}")
        print(f"  First data row:")
        print(df.iloc[0].to_string())
    except Exception as e:
        print(f"  ERROR: {e}")
        # Try with header on different row
        try:
            df = pd.read_excel(path, header=1, nrows=3)
            print(f"  Retry with header=1, columns:")
            for c in df.columns:
                print(f"    {repr(c)}")
        except Exception as e2:
            print(f"  Also failed with header=1: {e2}")
```

```python
import os, zipfile, pandas as pd
from pathlib import Path

raw_path = Path(r"C:\Users\jarem\OneDrive - London School of Economics\YEAR 2\1. Policy paper\policy-paper-repo\data\clean\outcome\Elections\raw")

# See what's in 2015
print("2015 folder contents:")
for f in os.listdir(raw_path / "2015"):
    print(f"  {f}")

print()

# Peek at whichever file looks like the gmina votes file
# Try the most likely filename — adjust if different
for fname in os.listdir(raw_path / "2015"):
    fpath = raw_path / "2015" / fname
    print(f"\n--- {fname} ---")
    try:
        if fname.endswith(".zip"):
            with zipfile.ZipFile(fpath) as zf:
                print("  ZIP contents:", zf.namelist())
                with zf.open(zf.namelist()[0]) as f:
                    df = pd.read_csv(f, sep=";", encoding="utf-8", nrows=2, dtype=str)
                    print("  Columns:", df.columns.tolist())
        elif fname.endswith((".xls", ".xlsx")):
            df = pd.read_excel(fpath, nrows=2, dtype=str)
            print("  Columns:", df.columns.tolist())
    except Exception as e:
        print(f"  Error: {e}")
```
